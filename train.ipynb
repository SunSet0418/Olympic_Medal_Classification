# Training Model (Linear Regression)

```python
import csv
import tensorflow as tf

data_x = []
data_y = []

def read_data():
    dataset = []
    with open('dataset/athlete_events.csv', 'r', encoding='utf-8') as f:

        reader = csv.reader(f)

        for row in reader:
            json = {}

            if row[3] == 'NA' or row[4] == 'NA' or row[5] == 'NA' or row[14] == 'NA':
                pass

            else:
                json['age'] = row[3]
                json['height'] = row[4]
                json['weight'] = row[5]
                json['medal'] = row[14]
                dataset.append(json)

        del dataset[0]

    return dataset


def set_train_data():

    datas = read_data()

    for row in datas:
        tmp = []

        tmp.append(float(row['age']))
        tmp.append(float(row['height']))
        tmp.append(float(row['weight']))

        data_x.append(tmp)

        if row['medal'] == 'Gold':
            data_y.append([1])
        elif row['medal'] == 'Silver':
            data_y.append([2])
        elif row['medal'] == 'Bronze':
            data_y.append([3])
        else:
            data_y.append(row['medal'])

    print("data len : " + str(len(data_x)))
    
set_train_data()

train_x = tf.placeholder(tf.float32)
train_y = tf.placeholder(tf.float32)

weight = tf.Variable(tf.random_normal([1, 30181]))
bias = tf.Variable(tf.random_normal([1]))

hypothesis = tf.add(tf.matmul(weight, data_x), bias)

#cost = tf.reduce_mean(tf.square(hypothesis - train_y))
cost = tf.reduce_mean(tf.nn.softmax_cross_entropy_with_logits_v2(logits=hypothesis, labels=train_y))

learning_rate = 0.01

train = tf.train.AdamOptimizer(learning_rate=learning_rate).minimize(cost)

saver = tf.train.Saver()

save_path = "./checkpoints/olympic_model.ckpt"

if __name__ == "__main__":
    
    with tf.Session() as sess:
        
        sess.run(tf.global_variables_initializer())

        for step in range(10001):

            cost_v, hyp_v, _ = sess.run([cost, hypothesis, train], feed_dict={train_x: data_x, train_y: data_y})
            if step % 100 == 0:
                print("STEP :",step, "Cost :", cost_v, "Hypothesis :",len(hyp_v))

        saver.save(sess, save_path)

```

# Training Model (Neural Net)

```python
import csv
import tensorflow as tf
import numpy as np

data_x = []
data_y = []

train_x = tf.placeholder(tf.float32, shape=[None, 3])
train_y = tf.placeholder(tf.float32, shape=[None, 1])

def read_data():
    dataset = []
    with open('dataset/athlete_events.csv', 'r', encoding='utf-8') as f:

        reader = csv.reader(f)

        for row in reader:
            json = {}

            if row[3] == 'NA' or row[4] == 'NA' or row[5] == 'NA' or row[14] == 'NA':
                pass

            else:
                json['age'] = row[3]
                json['height'] = row[4]
                json['weight'] = row[5]
                json['medal'] = row[14]
                dataset.append(json)

        del dataset[0]

    return dataset


def set_train_data():
    datas = read_data()

    for row in datas:
        tmp = []

        tmp.append(float(row['age']))
        tmp.append(float(row['height']))
        tmp.append(float(row['weight']))

        data_x.append(tmp)

        if row['medal'] == 'Gold':
            data_y.append([1])
        elif row['medal'] == 'Silver':
            data_y.append([2])
        elif row['medal'] == 'Bronze':
            data_y.append([3])
        else:
            data_y.append(row['medal'])

    x_shape = np.array(data_x)
    y_shape = np.array(data_y)

    print("data_X Shape : ", x_shape.shape)
    print("data_Y Shape : ", y_shape.shape)

set_train_data()

w1 = tf.Variable(tf.random_normal([3, 5]))
b1 = tf.Variable(tf.random_normal([5]))
l1 = tf.nn.relu(tf.add(tf.matmul(train_x, w1), b1))

w2 = tf.Variable(tf.random_normal([5, 5]))
b2 = tf.Variable(tf.random_normal([5]))
l2 = tf.nn.relu(tf.add(tf.matmul(l1, w2), b2))

w3 = tf.Variable(tf.random_normal([5, 1]))
b3 = tf.Variable(tf.random_normal([1]))
hypothesis = tf.add(tf.matmul(l2, w3), b3)

cost = tf.losses.mean_squared_error(labels=train_y, predictions=hypothesis)

learning_rate = 0.01

train = tf.train.AdamOptimizer(learning_rate=learning_rate).minimize(cost)

saver = tf.train.Saver()

save_path = "./checkpoints/olympic_nn_model.ckpt"

with tf.Session() as sess:
    
    sess.run(tf.global_variables_initializer())

    print("Leaning Start!")

    for step in range(10001):

        cost_v,  _ = sess.run([cost, train], feed_dict={train_x: data_x, train_y: data_y})
        if step % 100 == 0:
            print("STEP :",step, "Cost :", cost_v)
        
    saver.save(sess, save_path)
    
    result = sess.run([hypothesis], feed_dict={train_x : [[22.0, 185.0, 82.0]]})
    print(result)
            
    print("Finished")

```
